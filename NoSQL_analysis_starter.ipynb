### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query={
    "$and" : [{
                 'LocalAuthorityName':'City of London Corporation'
              },
              {
                   "RatingValue" : { "$gte" : '4'}
              }]
}
# Use count_documents to display the number of documents in the result
print('document count: ',establishments.count_documents(query))
# Display the first document in the results using pprint
docs=establishments.find(query)
pprint(docs[0])
```

```python
# Convert the result to a Pandas DataFrame
df=pd.DataFrame(docs)
# Display the number of rows in the DataFrame
print('number of rows: ',len(df))
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# This is a reference for Penang
'''{
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}'''

```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search=0.01
latitude=51.49014200
ulat=latitude+degree_search
llat=latitude-degree_search
#
longitude=0.08384000
ulon=longitude+degree_search
llon=longitude-degree_search
print(llat,ulat,llon,ulon)
print('')

query={
    "$and" : [
        {"RatingValue" :'5'},
        {'geocode.latitude':{'$gte':'%s'%llat,'$lte':'%s'%ulat}},
        {'geocode.longitude':{'$gte':'%s'%llon,'$lte':'%s'%ulon}},
                ]
}
docs=establishments.find(query)
sort=docs.sort('geocode.Hygiene',1)

# Print the results
for doc in sort:
    pprint(doc)
```

```python
# Convert result to Pandas DataFrame
docs=establishments.find(query)
sort=docs.sort('geocode.Hygiene',1)
df=pd.DataFrame(sort)
print(df.head())
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
establishments.count({
    "type": "LocalAuthority"
})
```

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
query=[{'$match':{"scores.Hygiene":0}},
       {'$group':{"_id":"$LocalAuthorityBusinessID",'count':{'$sum': 1}}},
       {'$sort': {'count': -1,'_id': 1 }}
      ]
# Print the number of documents in the result
docs=establishments.aggregate(query)
docs=list(docs)
print('number of documents: ',len(docs))
# Print the first 10 results
print('document count by LocalAuthorityBusinessID')
for doc in docs[:10]:
    print(doc)
```

```python
# Convert the result to a Pandas DataFrame
df=pd.DataFrame(docs)
# Display the number of rows in the DataFrame
print('number of rows: ',len(df))
# Display the first 10 rows of the DataFrame
df.head(10)
```
